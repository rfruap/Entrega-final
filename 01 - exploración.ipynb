```python

import os, pandas as pd

def load_table(base_name):
    csv_path = os.path.join('.', f'{base_name}.csv')
    xlsx_path = os.path.join('.', f'{base_name}.xlsx')
    if os.path.exists(csv_path):
        return pd.read_csv(csv_path), csv_path
    elif os.path.exists(xlsx_path):
        return pd.read_excel(xlsx_path), xlsx_path
    else:
        raise FileNotFoundError(f'No se encontró {base_name}.csv ni {base_name}.xlsx en la carpeta del notebook.')

import matplotlib.pyplot as plt

train, train_path = load_table('train')
print('Usando:', train_path, '| shape:', train.shape)
display(train.head(8))

try:
    test, test_path = load_table('test')
    print('Usando:', test_path, '| shape:', test.shape)
except FileNotFoundError:
    test = None
    print('test.csv/xlsx no encontrado (opcional).')
```

```python

# Distribución de clases (si existe RENDIMIENTO_GLOBAL)
import matplotlib.pyplot as plt
if 'RENDIMIENTO_GLOBAL' in train.columns:
    serie = (train['RENDIMIENTO_GLOBAL'].astype(str).str.strip().str.lower()
             .replace({'medio bajo':'medio-bajo','medio alto':'medio-alto'}))
    vc = serie.value_counts().sort_index()
    display(vc)
    plt.figure(); vc.plot(kind='bar')
    plt.title('Distribución de estudiantes por clase (train)')
    plt.xlabel('Clase'); plt.ylabel('Cantidad'); plt.tight_layout(); plt.show()
else:
    print('No se encontró la columna RENDIMIENTO_GLOBAL en train.')
```

```python

# Top-20 por una columna categórica de baja cardinalidad
def pick_column(df):
    for c in df.columns:
        if c == 'RENDIMIENTO_GLOBAL':
            continue
        if df[c].dtype == 'object' or df[c].nunique(dropna=True) <= 50:
            return c
    return None

col = pick_column(train)
print('Columna para Top-20:', col)
if col:
    vc = train[col].astype(str).str.strip().str.lower().value_counts().head(20)
    display(vc)
    import matplotlib.pyplot as plt
    plt.figure(); vc.plot(kind='bar')
    plt.title(f'Top 20 por {col} (train)')
    plt.xlabel(col); plt.ylabel('Cantidad'); plt.xticks(rotation=60); plt.tight_layout(); plt.show()
```
